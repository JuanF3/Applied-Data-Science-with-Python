### Question 1
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NHL** using **2018** data.

```python
#For all exercises you should import the librarias in the function according to coursera laboratory platform, but in this case i will not use a function for each question, I only import the libs once

#import libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
import re
```

```python
# Load the databases
NHL_df = pd.read_csv("assets/nhl.csv")
cities = pd.read_html("assets/wikipedia_data.html")[1]
cities = cities.iloc[:-1,[0,3,5,6,7,8]]
```

```python
#Cleaning NHL_df   
NHL_df = NHL_df[NHL_df['year'] == 2018]
NHL_df = NHL_df[['team','W','L']]
# Cleaning
def remover_star(team):
    if "*" in team:
        return re.findall("([\w\s]*)\*",team)[0]
    else:
        return team
    
NHL_df['team'] = NHL_df.team.apply(remover_star)
NHL_df['team_only'] = NHL_df.team.apply(lambda x: x.rsplit(None,1)[-1])
NHL_df
```

```python
#Changing some team names, because in the datasets 'cities'
NHL_df.loc[3, 'team_only'] = 'Maple Leafs'
NHL_df.loc[5, 'team_only'] = 'Red Wings'
NHL_df.loc[13, 'team_only'] = 'Blue Jackets'
NHL_df.loc[27, 'team_only'] = 'Golden Knights'

NHL_df = NHL_df[~NHL_df.W.str.contains('Division')]
NHL_df = NHL_df.reset_index(drop=True)

NHL_df.loc[12, 'team_only'] = 'Rangers Islanders Devils'
NHL_df.loc[14, 'team_only'] = 'Rangers Islanders Devils'
NHL_df.loc[15, 'team_only'] = 'Rangers Islanders Devils'

NHL_df.loc[24, 'team_only'] = 'Kings Ducks'
NHL_df.loc[26, 'team_only'] = 'Kings Ducks'

NHL_df.W = NHL_df.W.astype(float)
NHL_df.L = NHL_df.L.astype(float)

#we have the 28 teams!!
NHL_df
```

```python
#! Now Clean the Cities
def remove_bracket(team):
    if "[" in team:
        return re.findall("([\w\s]*)\[",team)[0]
    else:
        return team

cities['NHL'] = cities.NHL.apply(remove_bracket)

cities = cities [['Metropolitan area', 'Population (2016 est.)[8]','NHL']]

cities.rename(columns={'Metropolitan area':'city', 'Population (2016 est.)[8]':'population', 'NHL':'team_only'}, inplace=True)

cities['team_only'].unique()
```

```python
NHL_df = NHL_df.groupby(by='team_only')['W','L'].mean()
NHL_df.reset_index(inplace=True)
NHL_df
```

```python
merge_df = pd.merge(NHL_df,cities,how='left', on='team_only')
merge_df
```

```python
merge_df['win_loss_ratio'] = merge_df.W / (merge_df.W + merge_df.L)

 #raise NotImplementedError()
    
population_by_region = pd.to_numeric(merge_df['population'],errors='coerce') # pass in metropolitan area population from cities
win_loss_by_region = pd.to_numeric(merge_df['win_loss_ratio'],errors='coerce') # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]

assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    
stats.pearsonr(population_by_region, win_loss_by_region)[0]
```

### Question 2
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NBA** using **2018** data.

```python
#The same process of cleaning but with NBA

# Load the databases
NBA_df = pd.read_csv("assets/nba.csv")
cities = pd.read_html("assets/wikipedia_data.html")[1]
cities = cities.iloc[:-1,[0,3,5,6,7,8]]
NBA_df = NBA_df[NBA_df.year == 2018]
NBA_df = NBA_df [['team', 'W', 'L']]

# Cleaning names

NBA_df['team'] = NBA_df.team.apply(lambda team: re.findall('([\w\s]*)\*',team)[0] if '*' in team else team)
NBA_df['team'] = NBA_df.team.apply(lambda team: re.findall('([\w\s]*)\(',team)[0] if '(' in team else team)

#Extract the team only
NBA_df['team_only'] = NBA_df.team.apply(lambda x: x.rsplit(None,1)[-1])

NBA_df.loc[10, 'team_only'] = 'Knicks Nets'
NBA_df.loc[11, 'team_only'] = 'Knicks Nets'

NBA_df.loc[24, 'team_only'] = 'Lakers Clippers'
NBA_df.loc[25, 'team_only'] = 'Lakers Clippers'

NBA_df.loc[17, 'team_only'] = 'Trail Blazers'

NBA_df.W = NBA_df.W.astype(float)
NBA_df.L = NBA_df.L.astype(float)

#The following command is because we need to combine the teams that share the same city, grouping by mean
NBA_df = NBA_df.groupby(by='team_only')['W','L'].mean()

NBA_df.reset_index(inplace=True)

#! Now Clean the Cities
def remove_bracket(team):
    if "[" in team:
        return re.findall("([\w\s]*)\[",team)[0]
    else:
        return team

cities['NBA'] = cities.NBA.apply(remove_bracket)
cities = cities [['Metropolitan area', 'Population (2016 est.)[8]','NBA']]
cities.rename(columns={'Metropolitan area':'city', 'Population (2016 est.)[8]':'population', 'NBA':'team_only'}, inplace=True)

merge_df = pd.merge(NBA_df,cities,how='left', on='team_only')
merge_df['win_loss_ratio'] = merge_df.W / (merge_df.W + merge_df.L)
    

    #raise NotImplementedError()

population_by_region = pd.to_numeric(merge_df['population'],errors='coerce') # pass in metropolitan area population from cities
win_loss_by_region = pd.to_numeric(merge_df['win_loss_ratio'],errors='coerce') # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]

assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"


stats.pearsonr(population_by_region, win_loss_by_region)[0]
```

### Question 4
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NFL** using **2018** data.

```python
# Load the databases
NFL_df = pd.read_csv("assets/nfl.csv")
cities = pd.read_html("assets/wikipedia_data.html")[1]
cities = cities.iloc[:-1,[0,3,5,6,7,8]]
    
#Starting with cities in this time \\ cleaning

def remove_bracket(team):
    if "[" in team:
        return re.findall("([\w\s]*)\[",team)[0]
    else:
        return team

cities['NFL'] = cities.NFL.apply(remove_bracket)
cities = cities [['Metropolitan area', 'Population (2016 est.)[8]','NFL']]
cities.rename(columns={'Metropolitan area':'city', 'Population (2016 est.)[8]':'population', 'NFL':'team_only'}, inplace=True)

#Now with the NFL dataframe
NFL_df = NFL_df[NFL_df['year']==2018]
NFL_df = NFL_df[['team','W','L']]
    
#Cleaning the names of the team
NFL_df['team'] = NFL_df['team'].str.replace('*', '').str.replace('+', '')
NFL_df = NFL_df[~NFL_df['team'].str.contains('NFC|AFC')] # removing all who say NFC or AFC in the name

NFL_df['team_only'] = NFL_df['team'].apply(lambda x: x.rsplit(None, 1)[-1])
NFL_df.reset_index(inplace=True)
NFL_df.drop('index', axis=1, inplace=True)
    
NFL_df.loc[3, 'team_only'] = 'Giants Jets'
NFL_df.loc[19, 'team_only'] = 'Giants Jets'

NFL_df.loc[28, 'team_only'] = 'Rams Chargers'
NFL_df.loc[13, 'team_only'] = 'Rams Chargers'

NFL_df.loc[30, 'team_only'] = '49ers Raiders'
NFL_df.loc[15, 'team_only'] = '49ers Raiders'
    
NFL_df.W = NFL_df.W.astype(float)
NFL_df.L = NFL_df.L.astype(float)

NFL_df = NFL_df.groupby(by='team_only')['W','L'].mean()

NFL_df.reset_index(inplace=True)
merge_df = pd.merge(NFL_df,cities,how='left', on='team_only')
merge_df['win_loss_ratio'] = merge_df.W / (merge_df.W + merge_df.L)



#raise NotImplementedError()
    
population_by_region = pd.to_numeric(merge_df['population'],errors='coerce') # pass in metropolitan area population from cities
win_loss_by_region = pd.to_numeric(merge_df['win_loss_ratio'],errors='coerce') # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]
assert len(population_by_region) == len(win_loss_by_region), "Q4: Your lists must be the same length"
assert len(population_by_region) == 29, "Q4: There should be 29 teams being analysed for NFL"


stats.pearsonr(population_by_region, win_loss_by_region)[0]
```
